```python
# Load dataset
file_path = "/content/1730285881-Airbnb_Open_Data.xlsx"
df = pd.read_excel(file_path)

# -------------------------
# Basic Exploration
# -------------------------
print("Shape of dataset:", df.shape)
print("\nColumns:", df.columns)
print("\nMissing values:\n", df.isnull().sum())

# -------------------------
# Data Cleaning
# -------------------------
# Drop duplicates
df = df.drop_duplicates()

# Convert price & service fee to numeric (if not already)
df['price'] = pd.to_numeric(df['price'], errors='coerce')
df['service fee'] = pd.to_numeric(df['service fee'], errors='coerce')

# Fill missing reviews with 0
df['number of reviews'] = df['number of reviews'].fillna(0)

# -------------------------
# Descriptive Stats
# -------------------------
print("\nAverage Price:", df['price'].mean())
print("Median Price:", df['price'].median())
print("Top 5 Neighbourhoods:\n", df['neighbourhood group'].value_counts().head())

# -------------------------
# Visualizations
# -------------------------
plt.figure(figsize=(8,5))
sns.histplot(df['price'], bins=50, kde=True)
plt.xlim(0, 1000)  # limit to avoid extreme outliers
plt.title("Price Distribution")
plt.show()

plt.figure(figsize=(8,5))
sns.countplot(data=df, x='neighbourhood group', order=df['neighbourhood group'].value_counts().index)
plt.title("Listings per Neighbourhood Group")
plt.show()

plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='number of reviews', y='price', alpha=0.5)
plt.ylim(0, 1000)
plt.title("Price vs Number of Reviews")
plt.show()

# -------------------------
# Correlation Analysis
# -------------------------
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
df['review_month'] = df['last review'].dt.month
monthly_bookings = df['review_month'].value_counts().sort_index()

plt.figure(figsize=(8,5))
sns.lineplot(x=monthly_bookings.index, y=monthly_bookings.values, marker="o")
plt.title("Monthly Booking Patterns")
plt.xlabel("Month")
plt.ylabel("Number of Bookings (Reviews)")
plt.show()

# --- Booking by neighbourhood group ---
plt.figure(figsize=(8,5))
sns.countplot(x="neighbourhood group", data=df, order=df['neighbourhood group'].value_counts().index)
plt.title("Bookings by Neighbourhood Group")
plt.show()

# --- Average minimum nights (lead time proxy) ---
avg_lead_time = df['minimum nights'].mean()
print("Average booking lead time (minimum nights):", avg_lead_time)
```

```python
# --- Price distribution ---
plt.figure(figsize=(8,5))
sns.histplot(df['price'], bins=50, kde=True)
plt.xlim(0, 1000)
plt.title("Price Distribution")
plt.show()

# --- Price by neighbourhood ---
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x="neighbourhood group", y="price")
plt.ylim(0, 1000)
plt.title("Price by Neighbourhood Group")
plt.show()

# --- Relationship between price and reviews (proxy for satisfaction) ---
plt.figure(figsize=(8,5))
sns.scatterplot(x="price", y="number of reviews", data=df, alpha=0.5)
plt.xlim(0, 1000)
plt.title("Price vs Number of Reviews (Guest Interest)")
plt.show()
```

```python
# --- Room type popularity ---
plt.figure(figsize=(8,5))
sns.countplot(data=df, x="room type", order=df['room type'].value_counts().index)
plt.title("Guest Preference: Room Types")
plt.show()

# --- Average ratings by room type (proxy using review rate number) ---
avg_rating_by_room = df.groupby("room type")['review rate number'].mean()
print("Average rating by room type:\n", avg_rating_by_room)

# --- Amenities / House rules word analysis (optional advanced) ---
top_rules = df['house_rules'].dropna().str.split().explode().value_counts().head(20)
print("Most common words in house rules:\n", top_rules)

```

```python
# --- Host listing counts ---
plt.figure(figsize=(8,5))
sns.histplot(df['calculated host listings count'], bins=30, kde=False)
plt.title("Distribution of Host Listings")
plt.xlabel("Number of Listings per Host")
plt.show()

# --- Availability vs reviews (responsiveness proxy) ---
plt.figure(figsize=(8,5))
sns.scatterplot(x="availability 365", y="number of reviews", data=df, alpha=0.5)
plt.title("Host Availability vs Guest Reviews")
plt.show()

# --- High performing hosts (top by reviews) ---
top_hosts = df.groupby("host name")['number of reviews'].sum().sort_values(ascending=False).head(10)
print("Top 10 High Performing Hosts (by reviews):\n", top_hosts)

# --- Common complaints (from house_rules if available) ---
complaints = df['house_rules'].dropna().str.lower().str.contains("no")
complaint_ratio = complaints.mean() * 100
print(f"Percentage of listings with restrictive rules (possible complaints): {complaint_ratio:.2f}%")
```
